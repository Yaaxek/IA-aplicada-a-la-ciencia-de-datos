* Reasignamos el orden de columnas explícitamente para garantizar la estructura deseada.
* El resultado final `df` es tu base consolidada lista para análisis avanzado.

### **3a Pregunta:** ¿Cómo fueron las ventas a lo largo de los meses en 2023?


> #### <strong><font color="7b71ff" size=6>Prompt:</font></strong>

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de líneas que muestre las ventas totales mensuales de Zoop.

En "df", tenemos las siguientes columnas:
* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturación: calculada como el producto del precio unitario y la cantidad, sumado al costo de envío de cada venta.

Deseamos crear una visualización que muestre las ventas por mes. Primero, agrupa los datos por mes y luego crea una nueva columna llamada "mes" que contenga el nombre de cada mes, traduciendo los meses del índice mediante el siguiente diccionario. Utiliza la columna "fecha" para agrupar los datos y construir el gráfico. La columna "mes" solo debe usarse para cambiar las etiquetas del eje x.

```
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}
```


Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
import pandas as pd
import matplotlib.pyplot as plt

# Diccionario de traducción
meses = {
    'January': 'Ene',
    'February': 'Feb',
    'March': 'Mar',
    'April': 'Abr',
    'May': 'May',
    'June': 'Jun',
    'July': 'Jul',
    'August': 'Ago',
    'September': 'Sep',
    'October': 'Oct',
    'November': 'Nov',
    'December': 'Dic'
}

# Asegurar que la columna fecha esté en formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# Calcular facturación si no existe
df['facturacion'] = df['precio_unitario'] * df['cantidad'] + df['envio']

# Agrupar por mes
ventas_mensuales = (
    df
    .groupby(df['fecha'].dt.month_name())['facturacion']
    .sum()
    .reset_index()
)

# Renombrar columnas
ventas_mensuales.columns = ['mes', 'facturacion']

# Traducir nombres de meses
ventas_mensuales['mes'] = ventas_mensuales['mes'].map(meses)

# Ordenar correctamente por calendario
orden_meses = ['Ene','Feb','Mar','Abr','May','Jun','Jul','Ago','Sep','Oct','Nov','Dic']
ventas_mensuales['mes'] = pd.Categorical(
    ventas_mensuales['mes'],
    categories=orden_meses,
    ordered=True
)

ventas_mensuales = ventas_mensuales.sort_values('mes')

# =========================
# Crear gráfico de líneas
# =========================

plt.figure(figsize=(10, 5))

plt.plot(
    ventas_mensuales['mes'],
    ventas_mensuales['facturacion'],
    marker='o'
)

plt.title('Facturación mensual - Zoop')
plt.xlabel('Mes')
plt.ylabel('Facturación total')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
````

<font color="7b71ff" size=4> Comentarios: </font>

---

Cuando tenemos una cantidad mayor de períodos en nuestros datos, vale la pena recurrir a gráficos de líneas, que son ideales para series temporales. Dado que nuestros datos son mensuales, utilizar un gráfico de columnas, por ejemplo, reduciría la percepción de la secuencia de los datos. Las líneas ofrecen una mayor sensación de continuidad.

Siéntete libre de formular algunas hipótesis sobre los momentos de alta y baja en las compras.

Tenemos un DataFrame llamado "df" y queremos construir un gráfico de barras apiladas que muestre las ventas por trimestre en relación con el método de pago de Zoop.

* fecha: con las fechas de las compras en formato datetime (`aaaa-mm-dd`)
* facturacion: con los valores de las ventas
* metodo_pago: con el método de pago elegido en la compra

Deseamos crear una visualización que muestre las ventas por trimestre, agrupando los métodos de pago de cada trimestre correspondientes a cada compra.

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
# Asegurar que fecha esté en formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])

# Crear columna de trimestre
df['trimestre'] = df['fecha'].dt.to_period('Q')

# Agrupar por trimestre y método de pago
ventas_trimestre = (
    df
    .groupby(['trimestre', 'metodo_pago'])['facturacion']
    .sum()
    .reset_index()
)

# Convertir a formato pivot para gráfico apilado
ventas_pivot = ventas_trimestre.pivot(
    index='trimestre',
    columns='metodo_pago',
    values='facturacion'
)

# Convertir índice a string para mejor visualización
ventas_pivot.index = ventas_pivot.index.astype(str)
ventas_pivot
````

````python
# =========================
# Crear gráfico apilado
# =========================

ventas_pivot.plot(
    kind='bar',
    stacked=True,
    figsize=(10, 6)
)

plt.title('Facturación por trimestre y método de pago - Zoop')
plt.xlabel('Trimestre')
plt.ylabel('Facturación total')
plt.xticks(rotation=0)

plt.legend(title='Método de pago')
plt.tight_layout()
plt.show()
````

````python
def ventas_por_trimestre():
  import pandas as pd
  import matplotlib.pyplot as plt
  from matplotlib.ticker import FuncFormatter

  # --- Asegurar fecha como datetime ---
  df["fecha"] = pd.to_datetime(df["fecha"], errors="coerce")

  # --- Asegurar facturacion (si no existe) ---
  if "facturacion" not in df.columns:
      df["facturacion"] = (df["precio_unitario"] * df["cantidad"]) + df["envio"]

  # --- 1) Trimestre ---
  df["trimestre"] = df["fecha"].dt.to_period("Q")

  # --- 2) Agrupar por trimestre y método de pago ---
  ventas_trimestre = (
      df.groupby(["trimestre", "metodo_pago"])["facturacion"]
        .sum()
        .unstack(fill_value=0)
  )

  # --- Figura (10,6) con identidad Zoop ---
  fig, ax = plt.subplots(figsize=(10, 6))
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # --- Colores apilados según directriz ---
  palette = [BLANCO, GRIS_3, GRIS_1, GRIS_3]

  # --- 3) Gráfico de barras apiladas ---
  ventas_trimestre.plot(kind="bar", stacked=True, ax=ax, color=palette)

  # --- 4) Título ---
  ax.set_title("Ventas por Trimestre y Método de Pago en 2023", fontsize=18, color=GRIS_5)

  # Suprimir labels de ejes
  ax.set_xlabel("")
  ax.set_ylabel("")

  # xticks 0 grados, tamaño 12, color AZUL_1
  ax.tick_params(axis="x", rotation=0, labelsize=12, colors=AZUL_1)

  # yticks: "MXN$ valor M" sin decimales
  ax.yaxis.set_major_formatter(FuncFormatter(lambda x, pos: f"MXN$ {x/1e6:.0f} M"))
  ax.tick_params(axis="y", labelsize=12, colors=AZUL_1)

  # --- Quitar bordes ---
  for spine in ["top", "bottom", "left", "right"]:
      ax.spines[spine].set_visible(False)

  # --- Cuadrícula punteada en Y ---
  ax.grid(axis="y", linestyle=":", linewidth=1, alpha=0.7)

  # --- Leyenda ---
  ax.legend(title="Método de Pago por trimestre de 2023")

  plt.tight_layout()
  return plt
````

Utiliza las bibliotecas de visualización de datos de Python para crear esta visualización.


````python
def proporcion_cashback():
  import matplotlib.pyplot as plt
  import numpy as np

  # --- Normalizar valores de cashback a "Sí" / "No" ---
  cash = df["cashback"].astype(str).str.strip().str.lower()

  cash = cash.replace({
      "si": "sí", "sí": "sí", "s": "sí", "yes": "sí", "true": "sí", "1": "sí",
      "no": "no", "n": "no", "false": "no", "0": "no"
  })

  df_tmp = df.copy()
  df_tmp["cashback_norm"] = np.where(cash == "sí", "Sí", "No")

  # --- Calcular proporciones ---
  cashback_proporcion = df_tmp["cashback_norm"].value_counts()

  labels = ["Sí", "No"]
  values = [cashback_proporcion.get("Sí", 0), cashback_proporcion.get("No", 0)]
  colors = [AZUL_1, GRIS_3]  # Sí = AZUL_1, No = GRIS_3

  # --- Figura (6,6) con fondo Zoop ---
  fig, ax = plt.subplots(figsize=(6, 6))
  fig.patch.set_facecolor(GRIS_1)
  ax.set_facecolor(GRIS_1)

  # --- Gráfico de dona ---
  wedges, texts, autotexts = ax.pie(
      values,
      labels=labels,
      colors=colors,
      startangle=180,
      explode=(0, 0),
      autopct="%1.1f%%",
      pctdistance=0.8,
      wedgeprops={"width": 0.4}
  )

  # --- Ajustar etiquetas (mismo color que porción, tamaño 12) ---
  for text, color in zip(texts, colors):
      text.set_color(color)
      text.set_fontsize(12)

  # --- Ajustar porcentajes (BLANCO, tamaño 12) ---
  for autotext in autotexts:
      autotext.set_color(BLANCO)
      autotext.set_fontsize(12)

  # --- Título ---
  ax.set_title("¿Participan en el programa de cashback?", fontsize=18, color=GRIS_5)

  ax.set_aspect("equal")
  plt.tight_layout()
  return plt
````

````python
import pandas as pd
import numpy as np

# --- 1) Leer las 3 bases desde las URLs ya definidas: url_clientes, url_ventas, url_zoop_pay ---
df_clientes = pd.read_csv(url_clientes)
df_ventas   = pd.read_csv(url_ventas)
df_zoop_pay = pd.read_csv(url_zoop_pay)
````

````python
# --- 2) Estandarizar nombres de columnas + asegurar llaves ---
def standardize_columns(df: pd.DataFrame) -> pd.DataFrame:
    df = df.copy()
    df.columns = (
        df.columns.astype(str)
        .str.strip()
        .str.lower()
        .str.replace(r"\s+", "_", regex=True)
    )

    # Forzar id_compra si viene con alguna variante
    if "id_compra" not in df.columns:
        cand = [c for c in df.columns if ("id" in c and "compra" in c)]
        if cand:
            df = df.rename(columns={cand[0]: "id_compra"})

    # Forzar id_cliente si viene con alguna variante
    if "id_cliente" not in df.columns:
        cand = [c for c in df.columns if ("id" in c and "cliente" in c)]
        if cand:
            df = df.rename(columns={cand[0]: "id_cliente"})

    return df

df_clientes = standardize_columns(df_clientes)
df_ventas   = standardize_columns(df_ventas)
df_zoop_pay = standardize_columns(df_zoop_pay)
````

````python
# --- 3A) Merge LEFT ventas + clientes por id_compra (base = ventas) ---
if "id_compra" not in df_ventas.columns:
    raise KeyError(f"Falta 'id_compra' en df_ventas. Columnas: {list(df_ventas.columns)}")
if "id_compra" not in df_clientes.columns:
    raise KeyError(f"Falta 'id_compra' en df_clientes. Columnas: {list(df_clientes.columns)}")

df = df_ventas.merge(df_clientes, on="id_compra", how="left")

# --- 3B) Merge LEFT con Zoop Pay por id_cliente (zoop_pay no tiene id_compra) ---
if "id_cliente" in df.columns and "id_cliente" in df_zoop_pay.columns:
    df = df.merge(df_zoop_pay, on="id_cliente", how="left")
else:
    # Si no se puede unir, igual garantizamos que exista la columna zoop_pay
    if "zoop_pay" not in df.columns:
        df["zoop_pay"] = np.nan
````

````python
# --- 4) Tratar fecha ---
if "fecha" in df.columns:
    df["fecha"] = pd.to_datetime(df["fecha"], errors="coerce", dayfirst=True)
else:
    df["fecha"] = np.nan
````

````python
# --- 5) Orden EXACTO de columnas + crear faltantes con NaN ---
cols_order = [
    "id_compra", "fecha", "categoria", "precio_unitario", "cantidad", "envio", "metodo_pago",
    "id_cliente", "edad", "sexo_biologico", "ciudad", "estado", "region", "cashback", "nota", "zoop_pay"
]

for c in cols_order:
    if c not in df.columns:
        df[c] = np.nan

# Orden exacto y dejar extras al final
df = df[cols_order + [c for c in df.columns if c not in cols_order]]
````

````python
import numpy as np
import matplotlib.pyplot as plt

# --- Normalizar valores de zoop_pay a "Sí" / "No" ---
z = df["zoop_pay"].astype(str).str.strip().str.lower()

# Ajusta aquí si tu dataset trae otros valores (ej. "True/False", "1/0", "yes/no", etc.)
z = z.replace({
    "si": "sí", "sí": "sí", "s": "sí", "yes": "sí", "true": "sí", "1": "sí",
    "no": "no", "n": "no", "false": "no", "0": "no"
})

df_tmp = df.copy()
df_tmp["zoop_pay_norm"] = np.where(z == "sí", "Sí", "No")

# --- Conteo para el gráfico ---
counts = df_tmp["zoop_pay_norm"].value_counts()

labels = ["Sí", "No"]
values = [counts.get("Sí", 0), counts.get("No", 0)]
colors = [AZUL_1, GRIS_3]  # "Sí" azul, "No" gris

# --- Figura (6,6) con fondo gris ---
fig, ax = plt.subplots(figsize=(6, 6))
fig.patch.set_facecolor(GRIS_1)
ax.set_facecolor(GRIS_1)

# --- Donut chart ---
wedges, texts, autotexts = ax.pie(
    values,
    labels=labels,
    colors=colors,
    startangle=90,
    explode=(0, 0),          # sin explosiones
    autopct="%1.1f%%",
    pctdistance=0.8,         # porcentajes más hacia afuera
    labeldistance=1.05,      # etiquetas fuera (ligero)
    textprops={"fontsize": 12}
)

# --- Ajustar estilo de etiquetas (mismo color que su porción) ---
for t, c in zip(texts, colors):
    t.set_color(c)
    t.set_fontsize(12)

# --- Ajustar estilo de porcentajes (blanco, 12, negrita) ---
for at in autotexts:
    at.set_color(BLANCO)
    at.set_fontsize(12)
    at.set_fontweight("bold")

# --- Hacer rosca con ancho 0.4 ---
centre_circle = plt.Circle((0, 0), 0.6, fc=GRIS_1)  # radio interno = 1 - width = 0.6
ax.add_artist(centre_circle)

# --- Título ---
ax.set_title("Proporción de Pre-registro en Zoop Pay", fontsize=18, color=GRIS_5)

ax.set_aspect("equal")
plt.tight_layout()
plt.show()
````
